```python
import math
if line_pt1[0] != line_pt2[0]:
  slope = -(line_pt1[1] - line_pt2[1])/(line_pt1[0] - line_pt2[0])
  theta = math.degrees(math.atan(slope))
  if theta < 0:
    yaw = -90 - theta
  else:
    yaw = 90 - theta
  mat = cv2.getRotationMatrix2D((line_pt1[0],line_pt1[1]),yaw,1)
  affine_output = cv2.warpAffine(inside_output,mat,(1920,1440))
cv2_imshow(affine_output) 
```

```python
mask_dicts = glob.glob(root_path + mask_path + '*')
distance_id = 1
for mask_adress in mask_dicts:
  im = cv2.imread(mask_adress)
  imgray = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
  ret,imthresh = cv2.threshold(imgray,127,255,cv2.THRESH_BINARY)
  im_copy = imthresh.copy()
  contours , hierarchy = cv2.findContours(im_copy, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
  hull = cv2.convexHull(contours[0])
  hull_pts = np.array(hull)
  hull_pts = np.reshape(hull_pts,(hull.shape[0],hull.shape[2]))

  previous=[]
  max_hull_dis=0

  for hull_pt in hull_pts:
    if previous == []:
      previous=hull_pt
      initial=hull_pt  
    else:
      dis = (hull_pt[0]-previous[0])**2 + (hull_pt[1]-previous[1])**2
      if dis > max_hull_dis:
        max_hull_dis = dis     
        line_pt1=previous
        line_pt2=hull_pt
      previous = hull_pt

  dis = (previous[0]-initial[0])**2 + (previous[1]-initial[1])**2
  if dis > max_hull_dis:      
    line_pt1=previous
    line_pt2=initial

  contour_pts = np.array(contours)
  contour_pts = np.reshape(contour_pts , (contour_pts.shape[1] , contour_pts.shape[3]))
  for  index, contour_pt  in enumerate(contour_pts):
    if np.all(contour_pt == line_pt2):
      index_2 = index
    elif np.all(contour_pt == line_pt1):
      index_1 = index

  inside_pts = contour_pts[index_2 :index_1 + 1]
  inside_buffer = np.zeros(im.shape[:2], dtype=np.uint8)
  inside_pts = np.reshape(inside_pts , (inside_pts.shape[0], 1, inside_pts.shape[1]))

  inside_output = cv2.polylines(inside_buffer, [inside_pts], True, 255, 1)
  
  if line_pt1[0] != line_pt2[0]:
    slope = -(line_pt1[1] - line_pt2[1])/(line_pt1[0] - line_pt2[0])
    theta = math.degrees(math.atan(slope))
    if theta < 0:
      yaw = -90 - theta
    else:
      yaw = 90 - theta
    mat = cv2.getRotationMatrix2D((line_pt1[0],line_pt1[1]),yaw,1)
    affine_output = cv2.warpAffine(inside_output,mat,(1920,1440))

  dis_pts = list(zip(*np.where(affine_output >= 1)))
  dis_pts = np.array(dis_pts)
  dis_x_max = dis_pts.max(axis=0)[1]
  dis_x_min = dis_pts.min(axis=0)[1]
  dis_max = abs(dis_x_max - line_pt1[0])
  dis_min = abs(dis_x_min - line_pt1[0])
  if dis_max >= dis_min:
    dis_x = dis_x_max
    ans_pts = list(zip(*np.where(dis_pts == dis_x)))
    distance = dis_max
  else:
    dis_x = dis_x_min
    ans_pts = list(zip(*np.where(dis_pts == dis_x)))
    distance = dis_min
  dis_pt = dis_pts[ans_pts[0][0]]
  dis_output = cv2.line(affine_output, (line_pt1[0], dis_pt[0]), (dis_pt[1], dis_pt[0]), 255, thickness=1)
  if distance <= 26.11:
    cv2.putText(dis_output, "distance:"+str(distance)+"[pixel](excellent)",(10, 30), cv2.FONT_HERSHEY_PLAIN, 1.5, 255, 2)
  elif distance > 26.11 and distance <= 52.22:
    cv2.putText(dis_output, "distance:"+str(distance)+"[pixel](good)",(10, 30), cv2.FONT_HERSHEY_PLAIN, 1.5, 255, 2)
  else:
    cv2.putText(dis_output, "distance:"+str(distance)+"[pixel](cut)",(10, 30), cv2.FONT_HERSHEY_PLAIN, 1.5, 255, 2)

  cv2.imwrite(root_path+distance_path+str(distance_id)+'.png' , dis_output)
  distance_id += 1
  
```
